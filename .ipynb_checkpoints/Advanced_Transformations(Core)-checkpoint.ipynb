#**2. Import and Loading**

## Load Libraries & Functions

```python
#Load Libraries
import pandas as pd
import numpy as np
import os, json
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder 
```

## Load Data

```python
# Open saved file and preview again
Superhinfo = pd.read_csv("Data/superhero_info.csv", low_memory = False)
Superhinfo.info()
#Make a copy
df = Superhinfo
```

```python
# Open saved file and preview again
Superhpowers = pd.read_csv("Data/superhero_powers.csv", low_memory = False)
Superhpowers.info()
#Make a copy
df1 = Superhpowers
```

## Clean the Data

```python
#Duplicate count
## Display the number of duplicate rows in the dataset
print(f'There are {df.duplicated().sum()} duplicate rows.')
```

```python
#Duplicate count
## Display the number of duplicate rows in the dataset
print(f'There are {df1.duplicated().sum()} duplicate rows.')
```

```python
# Display the total number of missing values
print(f'There are {df.isna().sum().sum()} missing values.')
```

```python
# Display the total number of missing values
print(f'There are {df1.isna().sum().sum()} missing values.')
```

```python
#Check for categorical descrepencies
for col in df.select_dtypes(include='object').columns:
  print(df[col].value_counts())
```

```python
#Replace values
df["Hair color"].replace("black","Black", inplace=True)
df["Hair color"].replace("blond","Blond", inplace=True)
df["Hair color"].replace("Brownn","Brown", inplace=True)
```

```python
#REplace values on measurements
df['Measurements'] = df['Measurements'].str.replace("'",'"')
df['Measurements'] = df['Measurements'].str.replace(" cm","")
df['Measurements'] = df['Measurements'].str.replace(" kg","")
df['Measurements'] = df['Measurements'].apply(json.loads)
df['Measurements']
```

```python
#Split measurements
height_weight= df['Measurements'].apply(pd.Series)
height_weight
```

```python
# concat height_weight with original dataframe
df = pd.concat((df, height_weight), axis = 1)
df.head(2)
```

```python
#Convert to Numeric
df["Height"] = df["Height"].astype(float)
df["Weight"] = df["Weight"].astype(float)
```

```python
#drop measurements
df = df.drop(columns=['Measurements'])
df.head(2)
```

```python
#Split Hero|Plublisher
df[['Hero','Publisher']] = df['Hero|Publisher'].str.split('|',expand=True)
df.head()
```

```python
#Set column order
df = df.drop(columns=['Hero|Publisher'])
df = df[['Hero', 'Publisher', 'Gender', 'Eye color','Race','Hair color','Height', 'Skin color','Alignment','Weight']]
df.head()
```

```python
#Check for categorical descrepencies
for col in df1.select_dtypes(include='object').columns:
  print(df1[col].value_counts())
```

```python
#Split the values to only return unique
powers_split = df1['Powers'].str.split(",")
unique_powers = powers_split.explode().unique()
unique_powers
```

```python
#Convert array to dataframe and then explode
df1['Powers'] = pd.DataFrame(unique_powers)
exploded = df1.explode('Powers')
exploded[['hero_names','Powers']].head(5)
```

```python
## exploding the column of lists
## saving the unique values from the exploded column
cols_to_make = exploded['Powers'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    df1[col] = df1['Powers'].str.contains(col)
df1.head()
```

```python
#Drop Powers
df1 = df1.drop(columns=['Powers'])
df1.head(2)
```

```python
#Rename column
df1.rename(columns={'hero_names': 'Hero'}, inplace=True)
```

## Combine Data

```python
#Combine the data sets
df_main = pd.concat([df, df1], axis = 1)
df_main.head(30)
```

## Explore the Data

```python
#Compare the average weight of super powers who have Super Speed to those who do not.
#df_main.groupby('Hero').contains('Super Speed')['Weight'].mean()
df_main.groupby('Super Speed')['Weight'].mean()
```

```python
#What is the average height of heroes for each publisher?
df_main.groupby('Publisher')['Height'].mean()
```
