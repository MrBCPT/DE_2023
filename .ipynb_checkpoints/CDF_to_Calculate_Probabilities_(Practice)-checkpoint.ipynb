#**1. Goal**
- Use the normal cumulative distribution function with the mean and standard deviation of female height to calculate the probability that a female's height is?

#**2. Import and Loading**

## Load Libraries & Functions

```python
#Load Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

## Load Data

```python
# Open saved file and preview again
df_main = pd.read_csv("Data/weight_height_male_female.csv", low_memory = False)
df_main.info()
#Make a copy
df = df_main
```

```python
#Get the females height data
females = df.loc[ df['Gender'] == 'Female'].copy()
data = females['Height']
```

```python
low_end = 55.0
high_end = 56.0
mean = np.mean(data)
std = np.std(data)
min = np.min(data)
max = np.max(data)
```

```python
# Set the range and spacing for our normal curve
x_range = np.linspace(min, max,1000)
```

```python
#create the cdf for normal data with our mean and std
cdf = stats.norm.cdf(x_range, loc=mean, scale=std)
```

#**3. Plot and interpret Graphs**

```python
# Plot histogram with CDF for Female Height
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(x=data, stat='density',cumulative=True)
ax.plot(x_range,cdf)
ax.axvline(low_end, color= 'black')
ax.axvline(high_end, color ='black')
ax.axvspan(low_end, high_end, color ='yellow', zorder = 0)
ax.set_title('Cumulative Distribution of Female Height')
ax.axvline(mean,color='k',ls='--');
```

## ** Interpretation **
- Continuous distribution
- Negatively Skewed


```python
#Calculate the probability of a female having a height between 55 and 56 inches.
stats.norm.cdf(high_end, loc=mean,scale=std) - stats.norm.cdf(low_end,loc=mean, scale=std)
```

```python
#probability of the height being less than 60 inches
stats.norm.cdf(60, loc=mean,scale=std)
```

```python
#is greater than the mean
1 -stats.norm.cdf(mean, loc=mean,scale=std)
```
